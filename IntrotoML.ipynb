For matrix multiplication of $(n, k)$ and $(k, m)$ matrices, for every row, we need $k \times m$ multiplications and $(k-1) \times m$ additions. These operations are performed on $n$ rows. The total number of multiplications is $n \times k \times m$, and the total number of additions is $n \times (k-1) \times m$.

Ques 2. 

```python
Matrix_A = [[3,6,8,9],[0,4,7,2],[4,2,7,1]]
Matrix_B = [[8,5,3,7],[1,7,6,5],[9,3,7,2], [7,4,3,2]]

def matrix_multiply(A, B):
    result = [[0 for _ in range(len(B[0]))] for _ in range(len(A))]
    for i in range(len(A)):
        for j in range(len(B[0])):
            for k in range(len(A[0])):
                result[i][j]+=A[i][k]*B[k][j]
            
    return result

C = matrix_multiply(Matrix_A, Matrix_B)
print(C)
%timeit
```

```python
import numpy as np 
Mat_C = np.matmul(Matrix_A, Matrix_B)
print(Mat_C)
```

```python
import timeit
A = np.random.rand(100, 100)
B = np.random.rand(100, 100)

time_custom = timeit.timeit(lambda: matrix_multiply(A.tolist(), B.tolist()), number=100)
time_matmul = timeit.timeit(lambda: np.matmul(A, B), number=100)

print(f'Time to run matrix multiply by list is for 100 times {time_custom}')
print(f'Time to run matrix multiply by numpy is for 100 times {time_matmul}')
```

```python
import time
import numpy as np

def median_iterative(C):
    n = len(C)
    passes = n//2+1
    max_element = float('-inf')
    second_max_element = float('-inf')
    for _ in range(passes):
        max_element = max(C)
        C.remove(max_element)
        if len(C)%2==0:
            second_max_element = max(C)
            C.remove(second_max_element)
    return max_element if n%2!=0 else (max_element+second_max_element)/2
            

def median_sorted_method(C):
    sort_C = sorted(C)
    n = len(sort_C)
    if n%2 !=0:
        return sort_C[n//2]
    else:
        mid1 = sort_C[n//2-1]
        mid2 = sort_C[n//2]
        return (mid1+mid2)/2
    

array = np.random.uniform(1,50,10000000)

# start_time = time.time()
# iter_median = median_iterative(array.tolist())
# iter_time = time.time()-start_time

start_time = time.time()
sort_median = median_sorted_method(array.tolist())
sort_time = time.time()-start_time

start_time = time.time()
numpy_median = np.median(array) 
np_time = time.time()-start_time

# print(f'Median by iterative method is {iter_median} and time taken to run is {iter_time}s')
print(f'Median by sorted method is {sort_median} and time taken to run is {sort_time}s')
print(f'Median by numpy method is {numpy_median} and time taken to run is {np_time}s')
```

```python
x_value = np.arange(0.5,100,0.5)
y_f1 = f1(x_value)
y_f2 = f2(x_value)
y_f3 = f3(x_value)
y_f4 = f4(x_value)
y_f5 = f5(x_value)
y_f6 = f6(x_value)
y_f7 = f7(x_value)
```

Ques 12.

 Broadcasting describes how NumPy treats arrays with different shapes during arithmetic operations. Subject to certain constraints, the smaller array is broadcast across the larger array so that they have compatible shapes.

```python
a = [[2,0,8],[4,7,9]]
b = 3
c = a*b
print(c)
```

```python
a = np.array([[2,0,8],[4,7,9]])
b = np.array(3)
c = a*b
print(c)
```

We can clearly see the difference as we use numpy array it broadcasted smaller array for multiplication where as normal python just repeat three times

```python
def argmin(A):
    min_val = A[0]
    index = 0
    for i in range(len(A)):
        if A[i]<min_val:
            min_val = A[i]
            index = i
    return index
```

```python
Arr = [9, 5,1, 7, 2, 8]
ind = argmin(Arr)
print(ind)
ind_check=np.argmin(Arr)
print(ind_check)
```
